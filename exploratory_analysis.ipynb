```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузка данных
df = pd.read_csv("data/scoring.csv")
print(f"Размер датасета: {df.shape}")
df.head()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='default', data=df, palette='Set2')
plt.title('Распределение целевой переменной (Default)')
plt.xlabel('Дефолт (1 = Да, 0 = Нет)')
plt.ylabel('Количество записей')
plt.show()

# Вывод процентов
print(f"Процент дефолтов: {(df['default'].mean() * 100):.2f}%")
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='default', y='income', data=df, palette='Set2')
plt.title('Распределение дохода в зависимости от дефолта')
plt.xlabel('Дефолт')
plt.ylabel('Доход (тыс. руб.)')
plt.show()
```
